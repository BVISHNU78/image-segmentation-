```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread("/home/linux/Downloads/jonas-jaeken-b4BLUYKLzP4-unsplash.jpg")
plt.imshow(image)
pixel = image.reshape((-1,3))
pixel=np.float32(pixel)
ck=(cv2.TERM_CRITERIA_EPS+ cv2.TERM_CRITERIA_MAX_ITER,100,0.85)
k=10
retrval,labels,centre =cv2.kmeans(pixel,k,None,ck,5,cv2.KMEANS_PP_CENTERS)
center = np.uint8(centre)
segmented_data=center[labels.flatten()]
segemented_image=segmented_data.reshape(image.shape)
plt.imshow(segemented_image)
```
